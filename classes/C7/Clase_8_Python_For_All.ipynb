### Exercise 152: Number the Lines in a File
Create a program that reads lines from a file, adds line numbers to them, and then stores the numbered lines into a new file. The name of the input file will be read from the user, as will the name of the new file that your program will create. Each line in the output file should begin with the line number, followed by a colon and a space,
followed by the line from the input file.

```python
# Usuario entrega una ruta archivo de entrada
# Usuario entrega una ruta archivo de salida
# Entregamos archivo en ruta de salida con lineas numeradas
with open("files/people.txt", "r") as f:
    lines = f.readlines()
# print(lines)
numerated_lines = []
for i, line in enumerate(lines):
    numerated_lines.append(f"{i+1}, " + line)
numerated_content = "".join(numerated_lines)
print("".join(numerated_lines))
with open("files/people_with_number.txt", "w") as f:
    f.write(numerated_content)
```

```python
def limpiar_dataframe(path, output_path):
    df = pd.read_csv(path)
    df.columns = [column.strip() for column in df.columns]
    (
        df
        .drop(columns="1")
        .dropna()
        .drop_duplicates()
        .to_csv(output_path, index=False)
    )
```

```python
class Dog:
    def __init__(self, raza, edad, patas=4):
        self.raza = raza
        self.edad = edad
        self.extremidades = patas
        self.ladro = False

    def ladrar(self):
        self.ladro = True
        print("Woof")
    
    def perseguir_juguete(self):
        print("Salio corriendo")
```

```python
class Reader:
    def __init__(self, pandas, normal, path):
        self.pandas = pandas
        self.normal = normal
        self.path = path

    def estado(self):
        print(self.pandas, "es con pandas?")
        print(self.normal, "es normal?")
    
    def check_type(self):
        termination = self.path.split(".")[-1]
        if termination == "xlsx" or termination == "xls":
            return "excel"
        elif termination == "csv" or termination == "txt":
            return "csv"

    def read(self):
        file_type = self.check_type()
        if self.pandas is True:
            if file_type == "csv":
                return pd.read_csv(self.path)
            elif file_type == "excel":
                return pd.read_excel(self.path)
        if self.normal is True:
            with open(self.path, "r") as f:
                return f.readlines()
```
